### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

print(f"Ingreso total de la Tienda 1: ${ingreso_tienda1:,.2f}")
print(f"Ingreso total de la Tienda 2: ${ingreso_tienda2:,.2f}")
print(f"Ingreso total de la Tienda 3: ${ingreso_tienda3:,.2f}")
print(f"Ingreso total de la Tienda 4: ${ingreso_tienda4:,.2f}")
```

# 2. Ventas por categoría

```python
conteo_categorias_tienda1 = tienda.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
conteo_categorias_tienda2 = tienda2.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
conteo_categorias_tienda3 = tienda3.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
conteo_categorias_tienda4 = tienda4.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)

print("Cantidad de productos vendidos por categoría en la Tienda 1:")
print(conteo_categorias_tienda1)

print("\nCantidad de productos vendidos por categoría en la Tienda 2:")
print(conteo_categorias_tienda2)

print("\nCantidad de productos vendidos por categoría en la Tienda 3:")
print(conteo_categorias_tienda3)

print("\nCantidad de productos vendidos por categoría en la Tienda 4:")
print(conteo_categorias_tienda4)

print("\nCategoría más popular en la Tienda 1:", conteo_categorias_tienda1.index[0])
print("Categoría más popular en la Tienda 2:", conteo_categorias_tienda2.index[0])
print("Categoría más popular en la Tienda 3:", conteo_categorias_tienda3.index[0])
print("Categoría más popular en la Tienda 4:", conteo_categorias_tienda4.index[0])
```

# 3. Calificación promedio de la tienda


```python
promedio_calificacion_tienda1 = tienda['Calificación'].mean()
promedio_calificacion_tienda2 = tienda2['Calificación'].mean()
promedio_calificacion_tienda3 = tienda3['Calificación'].mean()
promedio_calificacion_tienda4 = tienda4['Calificación'].mean()

print(f"Calificación promedio de la Tienda 1: {promedio_calificacion_tienda1:.2f}")
print(f"Calificación promedio de la Tienda 2: {promedio_calificacion_tienda2:.2f}")
print(f"Calificación promedio de la Tienda 3: {promedio_calificacion_tienda3:.2f}")
print(f"Calificación promedio de la Tienda 4: {promedio_calificacion_tienda4:.2f}")
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tiendas = {'Tienda 1': tienda, 'Tienda 2': tienda2, 'Tienda 3': tienda3, 'Tienda 4': tienda4}

for nombre_tienda, datos_tienda in tiendas.items():
    # Calcular la cantidad de veces que se vendió cada producto
    conteo_productos = datos_tienda['Producto'].value_counts()

    # Identificar el producto más vendido (el primero en la lista ordenada)
    producto_mas_vendido = conteo_productos.index[0] if not conteo_productos.empty else "No hay ventas"
    cantidad_mas_vendido = conteo_productos.iloc[0] if not conteo_productos.empty else 0

    # Identificar el producto menos vendido (el último en la lista ordenada)
    producto_menos_vendido = conteo_productos.index[-1] if len(conteo_productos) > 1 else producto_mas_vendido
    cantidad_menos_vendido = conteo_productos.iloc[-1] if len(conteo_productos) > 1 else cantidad_mas_vendido

    print(f"\n--- {nombre_tienda} ---")
    print(f"Producto más vendido: {producto_mas_vendido} ({cantidad_mas_vendido} ventas)")
    print(f"Producto menos vendido: {producto_menos_vendido} ({cantidad_menos_vendido} ventas)")

    plt.figure(figsize=(10, 6))
    top_5_vendidos = conteo_productos.head(5)
    sns.barplot(x=top_5_vendidos.index, y=top_5_vendidos.values, palette="viridis")
    plt.title(f'Los 5 productos más vendidos en {nombre_tienda}')
    plt.xlabel('Producto')
    plt.ylabel('Cantidad de ventas')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    if len(conteo_productos) > 1:
        plt.figure(figsize=(10, 6))
        bottom_5_vendidos = conteo_productos.tail(5)
        sns.barplot(x=bottom_5_vendidos.index, y=bottom_5_vendidos.values, palette="magma")
        plt.title(f'Los 5 productos menos vendidos en {nombre_tienda}')
        plt.xlabel('Producto')
        plt.ylabel('Cantidad de ventas')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
    elif len(conteo_productos) == 1:
        print("Solo hay un producto vendido en esta tienda, por lo que es el más y el menos vendido.")
    else:
        print("No hay datos de ventas para esta tienda.")
```

# 5. Envío promedio por tienda

```python
costo_envio_promedio_tienda1 = tienda['Costo de envío'].mean()
costo_envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
costo_envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
costo_envio_promedio_tienda4 = tienda4['Costo de envío'].mean()

print(f"Costo de envío promedio de la Tienda 1: ${costo_envio_promedio_tienda1:,.2f}")
print(f"Costo de envío promedio de la Tienda 2: ${costo_envio_promedio_tienda2:,.2f}")
print(f"Costo de envío promedio de la Tienda 3: ${costo_envio_promedio_tienda3:,.2f}")
print(f"Costo de envío promedio de la Tienda 4: ${costo_envio_promedio_tienda4:,.2f}")
```

Visualización de los Resultados.

```python
tiendas = {'Tienda 1': tienda, 'Tienda 2': tienda2, 'Tienda 3': tienda3, 'Tienda 4': tienda4}
nombres_tiendas = list(tiendas.keys())
datos_tiendas = list(tiendas.values())

# 1. Gráfico de Barras: Ingreso Total por Tienda
ingresos_totales = [df['Precio'].sum() for df in datos_tiendas]
plt.figure(figsize=(10, 6))
sns.barplot(x=nombres_tiendas, y=ingresos_totales, palette="viridis")
plt.title('Ingreso Total por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingreso Total (COP)')
plt.ticklabel_format(style='plain', axis='y')
plt.tight_layout()
plt.show()

# 2. Gráfico de Pastel: Distribución de Categorías de Productos en la Tienda 1
conteo_categorias_tienda1 = datos_tiendas[0]['Categoría del Producto'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(conteo_categorias_tienda1, labels=conteo_categorias_tienda1.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title('Distribución de Categorías de Productos en la Tienda 1')
plt.tight_layout()
plt.show()

# 3. Gráfico de Dispersión: Precio vs. Calificación (combinado para todas las tiendas)
data_combinada = pd.concat(datos_tiendas, ignore_index=True)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Precio', y='Calificación', data=data_combinada, alpha=0.5)
plt.title('Relación entre Precio y Calificación del Cliente (Todas las Tiendas)')
plt.xlabel('Precio (COP)')
plt.ylabel('Calificación del Cliente')
plt.grid(True)
plt.tight_layout()
plt.show()

# 4. Gráfico de Barras: Distribución de Ventas por cada Tienda
plt.figure(figsize=(12, 8))

for i, (nombre_tienda, datos_tienda) in enumerate(tiendas.items()):
    conteo_categorias = datos_tienda['Categoría del Producto'].value_counts()
    plt.subplot(2, 2, i + 1)
    sns.barplot(x=conteo_categorias.index, y=conteo_categorias.values, palette="plasma")
    plt.title(f'Cantidad de Productos Vendidos por Categoría en {nombre_tienda}')
    plt.xlabel('Categoría del Producto')
    plt.ylabel('Cantidad de Ventas')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()

plt.suptitle('Comparativa de Ventas por Categoría entre Tiendas', y=1.02)
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Ajustar el layout para el título general
plt.show()
```

```python
# Gráfico de Dispersión: Permite Identificar la Concentración Geográfica de las Ventas de Cada Tienda
plt.figure(figsize=(15, 10))

for i, (nombre_tienda, datos_tienda) in enumerate(tiendas.items()):
    plt.subplot(2, 2, i + 1)
    sns.scatterplot(x='lon', y='lat', data=datos_tienda, alpha=0.6, s=50)
    plt.title(f'Ubicación de Ventas en {nombre_tienda}')
    plt.xlabel('Longitud')
    plt.ylabel('Latitud')
    plt.grid(True)
    plt.tight_layout()

plt.suptitle('Distribución Geográfica de Ventas por Tienda', y=1.02)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# Análisis básico de la relación geográfica con ingresos y calificaciones
for nombre_tienda, datos_tienda in tiendas.items():
    print(f"\n--- Análisis Geográfico Básico de {nombre_tienda} ---")
    print(f"Rango de Latitud: ({datos_tienda['lat'].min():.4f}, {datos_tienda['lat'].max():.4f})")
    print(f"Rango de Longitud: ({datos_tienda['lon'].min():.4f}, {datos_tienda['lon'].max():.4f})")
    print(f"Ingreso Total Estimado: ${datos_tienda['Precio'].sum():,.2f}")
    print(f"Calificación Promedio: {datos_tienda['Calificación'].mean():.2f}")
```
